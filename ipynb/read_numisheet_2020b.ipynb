```python
%load_ext autoreload
%autoreload 2
%matplotlib inline
%autosave 0
    
import logging
logging.basicConfig(format='%(asctime)s %(levelname)s:%(message)s', level=logging.DEBUG, datefmt='%I:%M:%S')

import os
import sys
import numpy as np
import pandas as pd
import sdata
import uuid
modulepath = os.getcwd()
modulepath
```

```python
workpath = os.path.abspath(os.path.join("..", "data", "numisheet_2020"))
workpath
os.listdir(workpath)
```

```python
df_meta = pd.read_csv(os.path.join(workpath, "analyze-NS-rawdata-attributes.csv"))
df_meta.drop_duplicates(inplace=True)
df_meta
```

```python
df_meta.rename(columns={"Variable":"name",
                       "Units":"unit",
                       "Description":"description"}, inplace=True)
df_meta
```

```python
df_meta["type"] = df_meta.Type.astype(str)
df_meta
```

```python
df_meta["dtype"] = df_meta.Type.astype(str)
df_meta.dtype = df_meta.dtype.map({"Real":"float", "Char":"str", "Character":"float", "nan":"float"})
df_meta
```

```python
df_meta["value"] = None
df_meta.drop(columns=["Type", "AxisLabel"], inplace=True)
df_meta
```

```python
df_meta.set_index(df_meta.name, inplace=True)
```

```python
df_meta.columns
df_meta = df_meta[['name', 'value', 'dtype', 'unit', 'description']]
df_meta
```

```python
df_meta.values.tolist()
```

```python
metadata = sdata.Metadata.from_list(df_meta.values.tolist())
metadata.df
```

```python
df_exp = pd.read_csv(os.path.join(workpath, "analyze-NS-rawdata.csv"))
df_exp.drop_duplicates(inplace=True)
df_exp
```

```python
name = "Numisheet_2020_TensionCompression"
description = "https://data.nist.gov/od/id/mds2-2202 file analyze-NS-rawdata.csv 2020-12-07"
data = sdata.Data(name=name, metadata=metadata, uuid=sdata.uuid_from_str(name), table=df_exp, description=description)
data
```

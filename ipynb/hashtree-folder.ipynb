```python
import json

import json

class TreeNode:
    def __init__(self, name, rootpath, parent=None):
        """
        :param name: Der Name des Knotens (z. B. Ordnername).
        :param rootpath: Der Root-Pfad des Baums.
        :param parent: Der Elternknoten.
        """
        self.name = name  # Der logische Name des Knotens (Subfolder-Name)
        self.parent = parent  # Elternknoten
        self.rootpath = rootpath.rstrip("/") if parent is None else f"{parent.rootpath}/{name}"  # Physischer Pfad
        self.children = {}  # Kinderknoten

    def __lt__(self, other):
        return self.name < other.name

    def get_child(self, path):
        path_list = path.strip("/").split("/")  # Pfad in eine Liste umwandeln
        current_node = self
        for part in path_list:
            if part in current_node.children:
                current_node = current_node.children[part]
            else:
                return None  # Pfad existiert nicht
        return current_node

    def add_path(self, logical_path):
        """
        Fügt einen Pfad hinzu. Der physische Pfad wird automatisch generiert.
        :param logical_path: Der logische Pfad im Baum (z. B. "folder1/folder2").
        """
        path_list = logical_path.strip("/").split("/")
        self.add_path_list(path_list)

    def add_path_list(self, path_list):
        if not path_list:
            return

        head, *tail = path_list
        if head not in self.children:
            self.children[head] = TreeNode(head, rootpath=self.rootpath, parent=self)

        self.children[head].add_path_list(tail)

    def full_name(self):
        if self.parent is None:
            return self.name  # Root-Knoten
        return f"{self.parent.full_name()}/{self.name}"

    def to_dict(self):
        """
        Konvertiert den Baum in ein verschachteltes Dictionary.
        """
        return {
            "name": self.name,
            "rootpath": self.rootpath,
            "children": {name: child.to_dict() for name, child in self.children.items()}
        }

    def to_json(self):
        """
        Exportiert den Baum in JSON-Format.
        """
        return json.dumps(self.to_dict(), indent=2)

    def print_tree(self, padding="", is_last=True, is_root=True):
        if is_root:
            print(f"{self.name} (Path: {self.rootpath})")
        else:
            branch = "└─" if is_last else "├─"
            print(f"{padding}{branch}{self.name} (Path: {self.rootpath})")
            padding += "   " if is_last else "│  "

        children = sorted(self.children.values())
        for count, child in enumerate(children):
            child.print_tree(padding, count == (len(children) - 1), is_root=False)

    def __repr__(self):
        return f"({self.name})"

    def __str__(self):
        return f"({self.name}|{self.rootpath})"

    def get_subtree_by_name(self, name):
        """
        Sucht im Baum nach einem Knoten mit dem angegebenen Namen.
        :param name: Der Name des gesuchten Knotens.
        :return: Der Knoten (TreeNode), wenn gefunden, sonst None.
        """
        if self.name == name:
            return self  # Aktueller Knoten passt
        
        for child in self.children.values():
            result = child.get_subtree_by_name(name)
            if result:
                return result  # Knoten gefunden
        
        return None  # Kein Knoten mit dem Namen gefunden

tree = TreeNode("root", rootpath="/tmp/project")
tree.add_path("mat1/a1/b2")
tree.add_path("mat1/a3")
tree.add_path("mat1/a2/b1/")
tree.add_path("mat1/a2/b4/")
tree.to_dict()
print(tree.rootpath, tree.name)
tree.print_tree()
```

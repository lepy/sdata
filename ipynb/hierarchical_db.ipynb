```python
import sqlite3

# Datenbankverbindung aufbauen
conn = sqlite3.connect('hierarchical.db')

# Tabelle erstellen
conn.execute('''
CREATE TABLE IF NOT EXISTS nodes (
    path TEXT PRIMARY KEY,
    parent_path TEXT,
    name TEXT,
    value TEXT,
    FOREIGN KEY (parent_path) REFERENCES nodes (path)
)
''')

# Daten einfügen
def insert_node(path, value):
    parts = path.strip("/").split("/")
    name = parts[-1]
    parent_path = "/".join(parts[:-1]) if len(parts) > 1 else ""

    conn.execute('''
    INSERT OR REPLACE INTO nodes (path, parent_path, name, value) 
    VALUES (?, ?, ?, ?)
    ''', (path, parent_path, name, value))
    conn.commit()

# Kinder eines Pfades abfragen
def get_children(parent_path):
    parent_path = parent_path.strip("/")
    cursor = conn.execute('SELECT path, value FROM nodes WHERE parent_path = ?', (parent_path,))
    return cursor.fetchall()

# Beispielverwendung
insert_node("/a/b/c", "Daten bei /a/b/c")
insert_node("/a/b/d", "Daten bei /a/b/d")
insert_node("/a/x/y", "Daten bei /a/x/y")

# Kinder von "/a/b" abrufen
print("Kinder von /a/b:")
for child in get_children("/a/b"):
    print(child)

```

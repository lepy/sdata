```python
class Process(Base):
    """Process class

    """

    def __init__(self, **kwargs):
        """

        :param inputs: {}
        :param kwargs:
        """
        self.default_attributes = []
        self.workflow_node_type = 'process'
        self.inputs = kwargs.get("inputs", {})
        self.outputs = kwargs.get("outputs", {})
        self.status = "init"
        super().__init__(**kwargs)
        self.set_default_attributes()
        
po = p = sdata_factory("MyProcessNode", Process)
p.__class__, p.__class__.__bases__
po, po.inputs
```

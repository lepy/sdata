```python
from sdata.process import ProcessData, ProcessNode

# Beispiel: Spezifische Subklassen
class SpecificInput(ProcessData):
    def __init__(self, **kwargs):
        kwargs['default_attributes'] = [{"name":"a", "value": None, "dtype":float, "label":"an a", "required":True},
                                        {"name": "b", "value": None, "dtype": int, "label": "an b", "required":True},
                                        ]
        super().__init__(**kwargs)

class SpecificOutput(ProcessData):
    pass

class SpecificProcessNode(ProcessNode):
    required_inputs = [SpecificInput]  # Definiert benötigte Inputs
    required_outputs = [SpecificOutput]  # Definiert erzeugte Outputs

    def run(self) -> Dict[str, ProcessData]:
        # Beispiel-Verarbeitung: Erzeuge Output basierend auf Inputs
        self.validate_inputs()  # Validation vor Run

        # Simuliere Verarbeitung
        processed_data = {k: v.metadata.to_dict() for k, v in self.inputs.items()}  # Assuming to_dict() exists
        output = SpecificOutput(name="Output1", metadata={"processed": processed_data})
        self.outputs["output1"] = output
        return self.outputs
```

```python

input_a = SpecificInput(name="input", attributes=[{"name": "a", "value": 1.2},
                                                  {"name": "b", "value": 3}])
# Gültiger Knoten
node = SpecificProcessNode(
    name="MyNode",
    inputs={"input1": SpecificInput(name="Input1", metadata={"data": "value"})}
)
try:
    node.validate_inputs()  # Erfolgreich
    outputs = node.run()  # Erzeugt Outputs
    print("Valid node ran successfully.")
except ValueError as e:
    print(f"Valid node error: {e}")
node
```

```python
# Ungültiger Knoten (fehlender Input)
invalid_node = SpecificProcessNode(name="InvalidNode", inputs={})
try:
    invalid_node.validate_inputs()  # ValueError: Fehlender erforderlicher Input: SpecificInput
except ValueError as e:
    print(f"Invalid node error (expected): {e}")

# Unvollständiger Input
incomplete_node = SpecificProcessNode(
    name="IncompleteNode",
    inputs={"input1": SpecificInput(name="Input1")}  # Keine Metadaten, defaults with None
)
try:
    incomplete_node.validate_inputs()  # ValueError: Input 'input1' (Input1) ist nicht vollständig.
except ValueError as e:
    print(f"Incomplete node error (expected): {e}")
```

```python
%load_ext autoreload
%autoreload 2
%matplotlib inline
#%matplotlib notebook
%autosave 0
    
import logging
logging.basicConfig(format='%(asctime)s %(levelname)s:%(message)s', level=logging.DEBUG, datefmt='%I:%M:%S')

import os
import sys
import numpy as np
import pandas as pd
import sdata
import sdata.sclass
```

```python
df = pd.DataFrame({"a": [1., 2.], "b": ["a", "b"], "c": [4, 5]})
print(df.dtypes)
df
```

```python
project = sdata.SUUID.from_name(class_name="Project", name="My_Project")
project
```

```python
column_metadata = sdata.Metadata(name="column_metadata")
column_metadata.add("a", unit="mm", description="Länge a", label="$a$", required=True, ontology="qudt:Length")

sdf = sdata.sclass.DataFrame(name="my dataframe with metadata", df=df, project=project,
                            column_metadata=column_metadata)
print(sdf)
sdf.cmdf
```

```python
sdf.mdf
```

```python
sdf.cmdf
```

```python
parquet_bytes = sdf.to_parquet(compression="zstd")
print(len(parquet_bytes))
parquet_bytes[:30]
```

```python
sdf2 = sdata.sclass.DataFrame.from_parquet_bytes(parquet_bytes)
sdf2.mdf
```

```python
sdf2.cmdf
```

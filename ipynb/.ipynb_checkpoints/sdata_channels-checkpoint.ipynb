```python
%load_ext autoreload
%autoreload 2
%matplotlib inline
%autosave 0
    
import logging
logging.basicConfig(format='%(asctime)s %(levelname)s:%(message)s', level=logging.WARNING, datefmt='%I:%M:%S')

import os
import sys
import numpy as np
import pandas as pd
import sdata
import sdata.iolib.isomme
import uuid
print("sdata v{}".format(sdata.__version__))
```

## create a Data object

```python
data = sdata.Data(name="data2", 
                  uuid="38b26864e7794f5182d38459bab85841", 
                  table=pd.DataFrame({"a":[1.1, 2, 3], "b":[3., 1., 4.]}))
data.metadata.add("my_key", 123, unit="m^3", description="a volume", label="V")
data.metadata.add("force", 1.234, unit="kN", description="x force", label="F_x")
data
```

```python
data.df
```

```python
iso = sdata.iolib.isomme.Isomme.from_sdata(data)
iso.channels
```

```python
iso.export(exportpath=f"/tmp/{iso.name}", createall=True)
```

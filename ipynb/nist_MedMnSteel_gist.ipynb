```python
datas = []
for url in urls:
    js = requests.get(url).text
    data = sdata.Data.from_json(js)
    datas.append(data)
```

```python
datas
```

```python
data.metadata.df
```

```python
fig, ax = plt.subplots(figsize=[13,9])
for data in datas:
    sr = data.metadata.get("Target Strain Rate")
    if data.name=="MedMnSteel_10mps_2_RD":
        continue
    x_name = "displacement"
    y_name = "force"
    x_name = "eps_tech"
    y_name = "sig_tech"
    df = data.df
    ax.plot(df[x_name], df[y_name], label=r"$\dot \varepsilon = {1.value}$ {1.unit} - {0}".format(data.name, sr), marker="o", ms=2)

ax.legend()

x_attr = data.metadata.get(x_name)
y_attr = data.metadata.get(y_name)

ax.set_xlabel(r"{0.label} [{0.unit}]".format(x_attr))
ax.set_ylabel(r"{0.label} [{0.unit}]".format(y_attr))
ax.axhline(0, alpha=.7, lw=1, c="k")
ax.axvline(0, alpha=.7, lw=1, c="k")
fig.tight_layout()
```

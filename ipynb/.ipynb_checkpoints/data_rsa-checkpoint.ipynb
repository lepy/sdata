```python
def encrypt_long(message, public_key, chunk_size=(public_key['n'].bit_length() // 8) - 1):
    chunks = [message[i:i+chunk_size] for i in range(0, len(message), chunk_size)]
    return [sdata.iolib.rsa.encrypt(chunk, public_key) for chunk in chunks]

def decrypt_long(cipher_chunks, private_key):
    return b''.join(sdata.iolib.rsa.decrypt(c, private_key) for c in cipher_chunks)
```

```python
c = sdata.iolib.rsa.encrypt(message, public_key)
c
```

```python
c2 = sdata.iolib.rsa.decrypt(c, private_key)
c2
```
